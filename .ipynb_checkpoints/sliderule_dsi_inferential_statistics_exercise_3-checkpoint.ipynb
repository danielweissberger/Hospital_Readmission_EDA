# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
sns.set()

def get_regression_line(x, y):

    slope, intercept = np.polyfit(x.values, y.values, 1)
    x = [min(x.values), max(x.values)]
    y = [slope*xi + intercept for xi in x]
    text_string = 'y =' + str(round(slope, 5)) + '*x + ' + str(round(intercept,5))
    return x, y, text_string


def plot_2d(x, y, title, axarr=False, reg=False):
    if not axarr:
        axarr=plt
        plt.title(title)
    else:
        axarr.set_title(title)
        
    axarr.scatter(x,y,alpha=.5)
    
    
    if reg:
        x,y,text_string = get_regression_line(x,y)
        axarr.plot(x,y,color='red')
        axarr.text(2,1,text_string,fontweight='bold')
    

f, axarr = plt.subplots(3, 2, figsize=(12, 15))
f.delaxes(axarr[2, 1])

plot_2d(ami['Number of Discharges'], ami['Excess Readmission Ratio'], 'AMI', axarr[0,0], True)
plot_2d(hf['Number of Discharges'], hf['Excess Readmission Ratio'], 'HF', axarr[0,1], True)
plot_2d(pn['Number of Discharges'], pn['Excess Readmission Ratio'], 'PN', axarr[1,0], True)
plot_2d(copd['Number of Discharges'], copd['Excess Readmission Ratio'], 'COPD', axarr[1,1], True)
plot_2d(knee['Number of Discharges'], knee['Excess Readmission Ratio'], 'KNEE', axarr[2,0], True)

f.text(-0.01, 0.5, 'Excess Readmission Ratio', va='center', rotation='vertical', fontsize = 12)


plt.tight_layout()
plt.show()
```

Looking at this data and applying a least squares fit to this, we see that PN and COPD show that increase discharge rate increases the Excess Readmission Ratio. Otherwise for other measures (hip-knee, ami and heart failure) we see the opposite relationship and increasing discharge rate seems to increase the ratio. The latter supports the study findings however how statistically signicant is this? The slopes showing evidence of the latter are very small, could it be that this is just noise?

```python
def draw_bs_pairs_linreg(x, y, size=1):
    """Perform pairs bootstrap for linear regression."""

    # Set up array of indices to sample from: inds
    inds = np.arange(len(x))

    # Initialize replicates: bs_slope_reps, bs_intercept_reps
    bs_slope_reps = np.empty(size)
    bs_intercept_reps = np.empty(size)
    bs_r_vals = np.empty(size)
    bs_p_vals = np.empty(size)
    bs_std_err = np.empty(size)
    
    # Generate replicates
    for i in range(size):
        bs_inds = np.random.choice(inds, size=len(inds))
        bs_x, bs_y = x[bs_inds], y[bs_inds]
        bs_slope_reps[i], bs_intercept_reps[i], bs_r_vals[i], bs_p_vals[i], bs_std_err[i] = stats.linregress(bs_x, bs_y)
        
    return bs_slope_reps, bs_intercept_reps, bs_p_vals

```

```python
bs_slope_reps, bs_intercept_reps, bs_p_vals = draw_bs_pairs_linreg(clean_hospital_read_df['Number of Discharges'].values, 
                                                        clean_hospital_read_df['Excess Readmission Ratio'].values, 1000)

slope_min = min(bs_slope_reps)
slope_max = max(bs_slope_reps)
idx_slope_min = np.where(bs_slope_reps==slope_min)[0][0]
idx_slope_max = np.where(bs_slope_reps==slope_max)[0][0]

print(slope_min, slope_max)
```

```python
fig, ax = plt.subplots(figsize=(14,7))
_ = plt.hist(bs_slope_reps, bins=50, density=True)
_ = plt.xlabel('slope')
_ = plt.ylabel('PDF')
plt.xticks( rotation=60 )

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14,7))
plot_2d(clean_hospital_read_df['Number of Discharges'], 
        clean_hospital_read_df['Excess Readmission Ratio'], 'OVERALL', False, False)

x=np.array([0,7000])

for i in range(len(bs_slope_reps)):
    _ = plt.plot(x, bs_slope_reps[i]*x + bs_intercept_reps[i], linewidth=.5, alpha=.1, color='black')
    

_ = plt.plot(x, bs_slope_reps[idx_slope_min]*x + bs_intercept_reps[idx_slope_min], linewidth=2, color='blue', 
             label='min: y='+str(slope_min)+'*x + ' + str(bs_intercept_reps[idx_slope_min]) )
_ = plt.plot(x, bs_slope_reps[idx_slope_max]*x + bs_intercept_reps[idx_slope_max], linewidth=2, color='red', 
             label='max: y='+str(slope_max)+'*x + ' + str(bs_intercept_reps[idx_slope_max]) )

_ = plt.legend()
```

# B) Hypothesis Testing

Lets take a look at the mean values mentioned in the report and perform some hypothesis testing

```python
discharge_high = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges']>1000]['Excess Readmission Ratio']
discharge_low = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges']<100]['Excess Readmission Ratio']
```

```python
mean_discharge_high = np.mean(discharge_high)
```

```python
mean_discharge_high
```

```python
mean_discharge_low = np.mean(discharge_low)
```

```python
mean_discharge_low
```

Are these results statistically significant:

    Ho: The means are the same
    Ha: The mean for discharge high is greater than discharge low
    Alpha: .1 or 10%

```python
diff_of_means = mean_discharge_low-mean_discharge_high
std_low = np.std(discharge_low)
std_high = np.std(discharge_high)
```

```python
diff_of_means
```

```python
#Calculate the new standard devation for difference in distributions
sigma = np.sqrt((std_low**2)/len(discharge_low) + (std_high**2)/len(discharge_high))
sigma
```

```python
z = ((diff_of_means  - 0)/(sigma))
print('z-value: ', z)
```

```python
p_val_from_z = (1 - stats.norm.cdf(z))*2
print('p-value: ', p_val_from_z)
```

Given the null hypothesis is true there is a very low chance of obtaining such a difference in means. So we can say with very high confidence that this difference of means is significant and they are not the same. The alternate hypothesis is supported.

Finally lets check the pearson corellation between the two variables with the following hypothesis

    Ho: There is a significant corellation between discharges and excess readmission rate
    Ha: There is no significant corellation
    Alpha: .1 or 10% 


```python
plot = sns.jointplot('Number of Discharges', 'Excess Readmission Ratio', 
                  data=clean_hospital_read_df, kind='reg', color='b', height=9)

_ = plot.annotate(stats.pearsonr)
```

So we see that there is a some corellation but once again such a small p-value makes the results practically and statistically insignificant. We must reject the null hypothesis in favor of the alternate, no significant corellation.
